```python
import matplotlib.pyplot as plt 
brand = ["oneplus","redmi","samsumg","apple","nokia"]
x= [45,66,25,58,76]
c= ["red","purple","green","pink","orange"]
ex= [ 0,0,0,0.1,0]
plt.pie(x,labels = brand,colors= c,explode = ex ,shadow = True,autopct = "%0.2f",startangle = 90)
plt.show()
```

```python
import pandas as pd
data = pd.read_excel("expense3.xlsx")
df = pd.DataFrame(data)
groupedby  = df.groupby("Payment Mode")["Amount"].sum()
print(groupedby)
plt.pie(groupedby.values,labels = groupedby.index,autopct = "%0.2f")
plt.show()
print(df)
```
