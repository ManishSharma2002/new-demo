```python
import matplotlib.pyplot as plt
y= [95,88,65,23,50]
y1= [45,66,55,23,80]
x= ["part1","part2","part3","part4","part5"]
#color = ["red","blue","green","yellow","orange"]
plt.plot(x,y,marker = "o", ls=":",color = "purple",label = "week1",alpha = 1)    #  used for marking = "D","*","^"  , ls stand for linestyle: {'-', '--', '-.', ':', '',
plt.plot(x,y1,marker = "*", ls="-",color = "darkgreen",label = "week2",alpha = 0.5)    # alpha using the transprancy 
plt.legend(loc=0)  #loc= 1,2,3,4

plt.show()
```
