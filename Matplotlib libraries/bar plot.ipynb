```python
import matplotlib.pyplot as plt
y= [95,88,65,23,100]
x= ["part1","part2","part3","part4","part5"]
color = ["red","blue","green","yellow","orange"]
plt.bar(x,y,color = color,edgecolor ="black")
plt.xlabel("part of episode",fontsize = 15)
plt.ylabel("populartity",fontsize = 15)
plt.title("popular of different part of episode",fontsize = 15)
plt.show()
```
