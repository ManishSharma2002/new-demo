```python
import pandas as pd
data  =  pd.read_excel("expense3.xlsx")
df= pd.DataFrame(data)
print(df)
df["Category"] = df["Category"].astype("str")
groupedby = df.groupby("Category").agg({"Amount":"sum"})
print(groupedby)
plt.step(groupedby.index ,groupedby["Amount"],where = "mid",marker = "D")
plt.show()
```
